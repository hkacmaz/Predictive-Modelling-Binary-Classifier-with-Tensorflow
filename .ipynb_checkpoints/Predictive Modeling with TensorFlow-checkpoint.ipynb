# Model

```python
#start to create model
```

```python
#libraries for model
import tensorflow as tf
from sklearn.model_selection import train_test_split
```

```python
'''x34,x35,x68 and x93 are object columns so will be as categorical columns, if we have 
every unique word as a list, will be easy to implement'''
```

```python
#get unique words as a list
x34_list = list(data_train.x34.unique())
x35_list = list(data_train.x35.unique())
x68_list = list(data_train.x68.unique())
x93_list = list(data_train.x93.unique())
```

```python
#create categorical data in the form of tensorflow like.
cat_col_x34 = tf.feature_column.categorical_column_with_vocabulary_list(key='x34',vocabulary_list=x34_list)
cat_col_x35 = tf.feature_column.categorical_column_with_vocabulary_list(key='x35',vocabulary_list=x35_list)
cat_col_x68 = tf.feature_column.categorical_column_with_vocabulary_list(key='x68',vocabulary_list=x68_list)
cat_col_x93 = tf.feature_column.categorical_column_with_vocabulary_list(key='x93',vocabulary_list=x93_list)
```

```python
'''in the case categorical, we have two other options, hash coding, one hot encoding, 
but as of I know all the unique words and the number is very limited I prefered categorical by the vocab list'''
```

```python
#we need to have dataframe w/o y column to train model, because our goal is to predict y values. 
#if we have y values in the training dataset, then we will have the problem is being called DATA LEAK
# which means target data is hidden in the training data, which we will never have in the real life implementation of model
# so avoiding DATA LEAK is very vital to have durable, stable models.
x_dataset_train = data_train.drop('y',axis=1)
```

```python
#to work better on dataframe to distinguish number columns and object columns
num_cols = list(x_dataset_train.select_dtypes(exclude=object).columns)

obj_cols = list(x_dataset_train.select_dtypes(include=object).columns)

num_cols_names = data_train.drop('y',axis=1).select_dtypes(exclude=object).columns
```

```python
# to have dictionary just for categorical data is going to ease our further step
obj_col_dict = {'x34':cat_col_x34,'x35':cat_col_x35,'x68':cat_col_x68,'x93':cat_col_x93}
```

```python
'''building tf.columns by combining number cols with categorical cols'''
feat_columns = []
```

```python
#categorical cols cannot be used directly so need to pass thru indicator 
#so we will use dictionary to speed up and simplify the process
for col in data_train.columns:
    if col in obj_col_dict:
        feat_columns.append(tf.feature_column.indicator_column(obj_col_dict[col]))

    elif col in num_cols: #is_numeric_dtype is pandas function
        feat_columns.append(tf.feature_column.numeric_column(col))
```

```python
# training and test data is going to be splitted

x_train_data = data_train.drop('y',axis=1)

y_label = data_train['y']
```

```python
#split data (train_test_split is imported from sklearn.model_selection)
X_train, X_test, y_train, y_test = train_test_split(x_train_data, y_label, test_size=0.2, random_state=101)
```

```python
'''time for input function, optimizer, maybe some additional adjustments on optimizer and final step before
training which is compiling the model'''
```

```python
#input func - shuffle is good for training to avoid baising in the data.
input_func = tf.estimator.inputs.pandas_input_fn(x=X_train,y=y_train,batch_size=50,num_epochs=100,shuffle=True)
```

```python
#optimizer
my_optimizer = tf.train.AdagradOptimizer(learning_rate=0.01)
```

```python
#limiting the gradient to avoid excessive jumps
my_optimizer = tf.contrib.estimator.clip_gradients_by_norm(my_optimizer,5.0)
```

```python
#compile model
model = tf.estimator.DNNClassifier(hidden_units=[100,200,50],optimizer=my_optimizer,feature_columns=feat_columns)
```

```python
'''I have chosed AdagradOptimizer rather than adamoptimizer, 
because adaptive gradient allows the learning rate to adapt based on parameters. 
Another advantage is that it basically eliminates the need to tune the learning rate.'''
```

```python
'''I haven't chosed adamoptimizer (Adam or adaptive momentum) because it could be hard to control 
momentum in the beginning of the model w/o precisely tuned'''
```

```python
'''time to train model with 20K steps'''
```

```python
#training
model.train(input_fn=input_func,steps=20000)
```

```python
#DON'T shuffle at all, because we already split the data and the order will be change.
pred_fn = tf.estimator.inputs.pandas_input_fn(x=X_test,batch_size=len(X_test),shuffle=False)
```

```python
pred_gen = model.predict(input_fn=pred_fn)
```

```python
predictions = list(pred_gen)
```
